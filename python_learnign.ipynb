## Types of Functions in Python

Functions are reusable blocks of code that perform a specific task. They help organize code, make it more readable, and promote code reuse. Here's an overview of different function types and their syntax in Python.

### 1. User-Defined Functions

These are functions that you write yourself to perform a specific action. They are defined using the `def` keyword.

#### Basic Function (no arguments, no return value)

```python
def greet():
    print("Hello, Python!")

# Call the function
greet()
```

#### Function with Positional Arguments

Arguments are values passed to the function when it's called. Positional arguments are matched based on their order.

```python
def add(a, b):
    print(f"The sum of {a} and {b} is {a + b}")

# Call the function with arguments
add(5, 3)
add(10, 20)
```

#### Function with a Return Value

The `return` statement is used to send a result back to the caller. If no `return` is used, the function implicitly returns `None`.

```python
def multiply(x, y):
    return x * y

# Call the function and store the result
result = multiply(4, 6)
print(f"The product is: {result}")

print(f"Another product: {multiply(7, 8)}")
```

#### Function with Default Arguments

Default values can be provided for parameters. If a value isn't passed for that parameter, the default value is used.

```python
def say_hello(name="Guest", greeting="Hello"):
    print(f"{greeting}, {name}!")

# Call with default arguments
say_hello()

# Call with custom name
say_hello("Alice")

# Call with custom greeting and name
say_hello("Bob", "Hi")
```

#### Function with Keyword Arguments

Arguments can be passed using their parameter names, allowing you to specify them in any order.

```python
def display_info(name, age):
    print(f"Name: {name}, Age: {age}")

# Call with positional arguments
display_info("Charlie", 30)

# Call with keyword arguments (order doesn't matter)
display_info(age=25, name="David")
```

#### Function with Arbitrary Positional Arguments (`*args`)

Use `*args` to allow a function to accept an arbitrary number of positional arguments. These arguments are received as a tuple.

```python
def sum_all(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print(f"Sum of 1, 2, 3: {sum_all(1, 2, 3)}")
print(f"Sum of 10, 20, 30, 40: {sum_all(10, 20, 30, 40)}")
```

#### Function with Arbitrary Keyword Arguments (`**kwargs`)

Use `**kwargs` to allow a function to accept an arbitrary number of keyword arguments. These arguments are received as a dictionary.

```python
def print_profile(**details):
    for key, value in details.items():
        print(f"{key.replace('_', ' ').title()}: {value}")

print_profile(name="Eve", age=28, city="New York")
print("\n---")
print_profile(product="Laptop", price=1200, brand="TechCo", in_stock=True)
```

### 2. Lambda Functions (Anonymous Functions)

Lambda functions are small, anonymous functions defined with the `lambda` keyword. They can take any number of arguments but can only have one expression. They are often used for short, single-expression operations.

```python
# A lambda function to add two numbers
add_lambda = lambda x, y: x + y
print(f"Lambda sum: {add_lambda(10, 5)}")

# A lambda function to check if a number is even
is_even = lambda num: num % 2 == 0
print(f"Is 7 even? {is_even(7)}")
print(f"Is 8 even? {is_even(8)}")
```

### 3. Built-in Functions

Python comes with many pre-defined functions that are always available for use. Examples include `print()`, `len()`, `type()`, `sum()`, `max()`, `min()`, `input()`, etc.

```python
my_list_data = [10, 50, 20, 80, 30]
my_string_data = "Hello Python"

print(f"Length of list: {len(my_list_data)}")
print(f"Max value in list: {max(my_list_data)}")
print(f"Type of my_string_data: {type(my_string_data)}")

# Example of input()
# user_name = input("Enter your name: ")
# print(f"Welcome, {user_name}!")
```

### 4. Generator Functions

Generator functions are a special type of function that returns an iterator that can be iterated upon. They use the `yield` keyword instead of `return` to produce a sequence of results one at a time, pausing execution and resuming from where it left off. This is memory-efficient for large sequences.

```python
def count_up_to(n):
    i = 1
    while i <= n:
        yield i
        i += 1

# Create a generator object
counter = count_up_to(5)

# Iterate through the generator
print("Counting:")
for num in counter:
    print(num)

# Generators can be exhausted. To reuse, create a new generator object.
print("Counting again:")
counter_new = count_up_to(3)
print(list(counter_new)) # Convert to list to see all values at once
```

### 5. Methods

Methods are functions that belong to a class and operate on an object (an instance of that class). They are defined similarly to regular functions but include `self` as their first parameter, which refers to the instance of the class.

```python
class Dog:
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    # This is a method
    def bark(self):
        print(f"{self.name} says Woof!")

    # Another method
    def get_info(self):
        return f"{self.name} is a {self.breed}."

# Create an object (instance) of the Dog class
my_dog = Dog("Buddy", "Golden Retriever")

# Call the methods on the object
my_dog.bark()
print(my_dog.get_info())
```
